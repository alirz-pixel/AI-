```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 성적 분석 예제
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("data/score.csv")  # data frame
df
```

```python
df.info()
```

```python
df.describe()  # 통계 정보 (중에서 숫자로 된 것들)
```

```python
subjects = [ '국어', '영어', '수학', '과학' ]  # 리스트 : 배열
```

```python
# df에 총점 컬럼 추가하기  -->  df['추가할_컬럼'] = 계산식
df['총점'] = df[subjects[0]] + df[subjects[1]] + df[subjects[2]] + df[subjects[3]]
df
```

```python
# df의 칼럼들을 행 방향으로 모두 더하기  -->  df['원하는_칼럼명'].sum()
df[ subjects ].sum()
```

```python
# df의 칼럼들을 열 방향으로 모두 더하기  -->  df['원하는_칼럼명'].sum( axis = 1 )
df[ subjects ].sum( axis=1 )
```

```python
df['총점'] = df[ subjects ].sum( axis=1 )  # axis=0 (디폴트:행), axis=1 (열)
df
```

```python
# 평균 추가하기
df['평균'] = df[ subjects ].mean( axis=1 )
df
```

```python
# 정렬 : 평균은 내림차순으로 정렬하고, (평균이 같을 경우) 이름은 오름차순으로 출력
# ascending : False (디폴트:내림차순)   ||   True (오름차순)

sorted_df = df.sort_values( ['평균', '이름' ], ascending=[ False, True ])
sorted_df
```

```python
# 그래프를 그릴 때 한글깨짐 방지
# 원인 : 디폴트 글꼴이 영어이기 때문
# 해결 : 한글글꼴로 변경

import matplotlib
matplotlib.rc('font', family='Malgun Gothic')
```

```python
# 시각화 : 정렬된 평균 값들을 막대그래프로 그리기
sorted_df['평균'].plot(kind='bar')
```

```python
sorted_df.index = sorted_df['이름']
sorted_df
```

```python
sorted_df['평균'].plot(kind='bar', figsize=(12, 8))
```

```python
sorted_df[subjects].plot(kind='bar', figsize=(12,8))
plt.savefig('image1.jpg')  # 현재 그래프 저장하기
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/titanic.xls")
df


# ML, DL의 결과는 총 가지로 나뉜다.
# (분류) : 이진분류[2개], 다중분류[여러개]    || 회귀[값]


## feature selection : 특성 선택
##### X
# pclass : 선실등급(1,2,3), 1등실이 가장 좋음,  범주형 데이터 -> (순서형), object
# sex : 성별(female:여성, male:남성),           범주형 데이터 -> (명목형), object
# age : 나이,  수치형 데이터,                   수치형 데이터 -> (연속형), float64
# sibsp : 부부형제자매의_수,                    수치형 데이터 -> (이산형), int64
# parch : 부모자식_수,                          수치형 데이터 -> (이산형), int64
# fare : 운임,                                  수치형 데이터 -> (연속형), float64

##### y
# 정답 : survived(0:사망, 1:생존), 이진분류
```

```python
import warnings
warnings.filterwarnings('ignore') # python warning 무시
```

```python
# 시각화
plt.figure(figsize=(12, 6)) # 그래프의 사이즈 바꾸기

# 그래프 2개가 나오게 하기
# subplot(행 열 번째수)
plt.subplot(121)   
df[ 'survived' ].value_counts().plot.pie(explode=[0, 0.1], autopct="%.2f%%")
plt.subplot(122)
sns.countplot( 'survived', data=df )

plt.show()
```

```python
# 시각화2 : 막대그래프에 값을 출력하기 (countplot 에 값들을 출력)
plt.figure(figsize=(12, 6)) # 그래프의 사이즈 바꾸기

# 그래프 2개가 나오게 하기
# subplot(행 열 번째수)
plt.subplot(121)   
df[ 'survived' ].value_counts().plot.pie(explode=[0, 0.1], autopct="%.2f%%")

plt.subplot(122)
graph = sns.countplot( 'survived', data=df )
for p in graph.patches: # 막대부분 = patches
    graph.annotate( f"{p.get_height()}", (p.get_x()+0.35, p.get_height()+5) )

plt.show()
```

```python
# 연령대 그래프 : 분포도 (히스토그램)
df['age'].hist(bins=20, grid=False, figsize=(18, 8))
```

```python
# 성별과 생존의 관계 출력
df.groupby(['sex', 'survived'])['survived'].count()
```

```python
df[ ['sex', 'survived'] ].groupby(['sex']).mean()
```

```python
# 남녀별 생존 그래프
plt.figure(figsize=(18,8))

sub1 = plt.subplot(121)
df[ ['sex', 'survived'] ].groupby(['sex']).mean().plot.bar(ax=sub1)

plt.subplot(122)
sns.countplot('sex', hue='survived', data=df) # 남여별 생존인원 수

plt.show()
```
